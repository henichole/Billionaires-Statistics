```python
import pandas as pd

# Read the 'ImdbTitleRatings' CSV file into a DataFrame
imdb_title_rating = pd.read_csv('ImdbTitleRatings.csv')

# Display the first few rows of the dataset
print(imdb_title_rating.head())

# Display the data types and non-null values of each column
print(imdb_title_rating.info())
```

```python
# Summary statistics for numerical columns
print(imdb_title_rating.describe())
```

```python
import matplotlib.pyplot as plt

# Plotting a histogram of 'averageRating'
plt.hist(imdb_title_rating['averageRating'], bins=20, color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Average Ratings')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot for 'averageRating' vs 'numVotes'
plt.scatter(imdb_title_rating['averageRating'], imdb_title_rating['numVotes'], alpha=0.5)
plt.xlabel('Average Rating')
plt.ylabel('Number of Votes')
plt.title('Correlation between Average Rating and Number of Votes')
plt.show()

```
