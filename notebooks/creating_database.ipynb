```python
import pandas as pd

name=pd.read_csv('ImdbName.csv')
title_basics=pd.read_csv('ImdbTitleBasics.csv', low_memory=False)
title_crew=pd.read_csv('ImdbTitleCrew.csv')
title_principals=pd.read_csv('ImdbTitlePrincipals.csv')
title_ratings=pd.read_csv('ImdbTitleRatings.csv')
```

```python
print(name.duplicated().sum())

print(title_basics.duplicated().sum())
print(title_crew.duplicated().sum())

print(title_principals.duplicated().sum())
print(title_ratings.duplicated().sum())

```

```python
print(name.isnull().sum())

print(title_basics.isnull().sum())
print(title_crew.isnull().sum())

print(title_principals.isnull().sum())
print(title_ratings.isnull().sum())

```

```python
import sqlite3
conn = sqlite3.connect('moviedb')
name.to_sql('name', conn, if_exists='replace', index=False)
title_basics.to_sql('basics', conn, if_exists='replace', index=False)
title_crew.to_sql('crew', conn, if_exists='replace', index=False)
title_principals.to_sql('principals', conn, if_exists='replace', index=False)
title_ratings.to_sql('ratings', conn, if_exists='replace', index=False)
```
